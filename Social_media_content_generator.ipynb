```python
# ✅ Updated Gradio App: LLaMA + Stable Diffusion Poster Generator

import gradio as gr
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image
from collections import Counter

# --------- Setup LLaMA Model ---------
model_name = "meta-llama/Llama-2-7b-chat-hf"
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
)

tokenizer = AutoTokenizer.from_pretrained(model_name, token=access_token)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto",
    token=access_token
)
text_generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    pad_token_id=tokenizer.eos_token_id
)


# --------- Setup Stable Diffusion ---------
sd_pipe = StableDiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-2",
    torch_dtype=torch.float16,
    use_auth_token=access_token
).to("cuda")

# --------- System Prompt ---------
system_prompt = """
You are a highly skilled Social Media Marketing Expert. You generate a very good prompt to feed into the poster generation model which will help the model to generate a visually stunning poster.
"""

# --------- Utility Function to Remove Repeats ---------
def remove_repeats(text):
    lines = text.split("\n")
    line_counts = Counter(lines)
    return "\n".join([line for line in lines if line_counts[line] < 3])

# --------- Image Generation ---------
def generate_image_from_prompt(prompt):
    with torch.autocast("cuda"):
        image = sd_pipe(prompt, guidance_scale=7.5).images[0]
    return image

# --------- Main Agent Function ---------
def social_media_agent(user_input):
    # Step 1: Generate poster prompt text
    prompt = f"<s>[INST] <<SYS>>{system_prompt}<</SYS>> {user_input} [/INST]"
    response = text_generator(prompt, max_new_tokens=300, do_sample=True, temperature=0.7, top_p=0.9)
    full_text = response[0]['generated_text']
    poster_text = full_text.split("[/INST]")[-1].strip() if "[/INST]" in full_text else full_text.strip()
    poster_text = remove_repeats(poster_text)

    # Step 2: Use 1st paragraph or sentence as image prompt
    short_prompt = poster_text.split("\n")[0][:180].replace("**", "").strip()
    image = generate_image_from_prompt(short_prompt)

    return poster_text, image

# --------- Gradio Interface ---------
gr.Interface(
    fn=social_media_agent,
    inputs=gr.Textbox(lines=4, label="Ask the Social Media Expert"),
    outputs=[
        gr.Textbox(label="Expert Advice & Poster Prompt"),
        gr.Image(label="Generated Poster Visual")
    ],
    title="📣 Social Media Marketing Expert + AI Poster Generator",
    description="Describe your campaign or product. Get marketing ideas + a sample poster!"
).launch(share=True)
```

```python
def generate_reel_script(product_description):
    prompt = (
        f"Write a sassy, Gen Z-style two minutes Instagram Reel script. Use emojis, slang, and a flirty, confident tone. Also guide the influencer on how to speak in the reel. "
        f"Talk directly to the audience like their bestie. Make it hype and fun with a cheeky call-to-action and a hashtag. "
        f"Product: {product_description}\n"
        f"Example: 'Babes 😍 this face mist is magic ✨ One spritz and you're glowing 💧 Snag it before it sells out 🔥'\n"
        f"Write something similar:"
    )
    output = text_generator(prompt, max_new_tokens=300, temperature=0.95, do_sample=True)
    full_text = output[0]['generated_text']
    script = full_text.split("Write something similar:")[-1].strip()
    return script

# Gradio Interface
iface = gr.Interface(
    fn=generate_reel_script,
    inputs=gr.Textbox(label="Enter your product description"),
    outputs=gr.Textbox(label="Generated Instagram Reel Script"),
    title="Gen Z Instagram Reel Script Generator",
    description="Get a hype, flirty 30-sec Reel script in Gen Z style with emojis and slang 🎉"
)

iface.launch()
```
